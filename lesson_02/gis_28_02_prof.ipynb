## For Loops

````python
pokemons = ['Bulbasaur', 'Ivysaur', 'Venusaur', 'Charmander', 'Charmeleon', 'Charizard', 'Squirtle']
````

````python
for pokemon in pokemons:
    print(f"{pokemon} is {len(pokemon)} letters long")
````

````python
projections = ["aitoff", "hammer", "lambert", "mollweide"]
````

````python
for proj in projections:
    print(proj)
    print(proj.title())
````

````python
for projection in projections:
    plt.figure(figsize=(10, 10))
    plt.subplot(projection=projection)
    plt.title(projection.title())
    plt.grid(True)

    plt.savefig(f'img/projections/{projection}.png')  
    plt.savefig(f'pdf/{projection}_projection.pdf')  

    plt.show()
````

### Handy trick:  Add an index to your list with `enumerate()` 

````python
enumerate(projections)
````

````python
for index, projection in enumerate(projections):
    print(f'index number {index} contains value {projection}')
````

````python
for index, projection in enumerate(projections):
    plt.figure(figsize=(10, 10))
    plt.subplot(projection=projection)
    plt.title(f'Map Number {index+1}\n{projection.title()} Projection\n')
    plt.grid(True)
````

````python
for i in range(len(projections)):
    print(i)
````

### Common Trick:  Building Up New Lists with `append()`

````python
crazy_list = []

for index, projection in enumerate(projections):
    crazy_list.append(projection.title() * index)
````


### Looping Over a Dictionary

> In Python 3.6 and beyond, **dictionaries are ordered data structures**, which means that they **keep their elements in the same order in which they were introduced**.

> This is a relatively new feature of Python’s dictionaries, and it’s a very useful one. But if you’re writing code that is supposed to be run in different Python versions, then you must not rely on this feature, because it can generate buggy behaviors.

> Another important feature of **dictionaries** is that they **are mutable data structures**, which means that you can add, delete, and update their items. It’s worth noting that this also means that they can’t be used as keys to other dictionaries, as they are not hashable objects. Source: https://realpython.com/iterate-through-dictionary-python/

````python
# Dictionaries
d = {'first_name': 'Snoop', 'last_name': 'Dogg', 'profession':'singer'}
````

#### Iterate through Keys

````python
for key in d.keys():
    print(key)
````

#### Iterate Through Values

````python
for value in d.values():
    print(value)
````

#### or if you have the key use it to get the value

````python
for key in d.keys():
    print(d[key])  
````

#### Like `enumerate()` you can unpack the key and value with `.items()`

````python
for key, value in d.items():
    print(key, ': ', value)
````

#### Now let's try and use f-strings to add thousands separator in the Population values

````python
"a sentence about a's about bananas.".replace('a','x')
````

````python
for key, value in d.items():
    print(f"His {key.replace('_',' ')} is {value}.")
````
