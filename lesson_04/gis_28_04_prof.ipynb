## `pandas.to_datetime`
> This function converts a scalar, array-like, Series or DataFrame/dict-like to a pandas datetime object.

[`https://pandas.pydata.org/docs/reference/api/pandas.to_datetime.html`](https://pandas.pydata.org/docs/reference/api/pandas.to_datetime.html)

```python
police_incidents_pdf['Date Occurred'] = pd.to_datetime(police_incidents_pdf['Date Occurred'])
```

```python
police_incidents_pdf.info()
```

```python
police_incidents_pdf['Date Occurred'].value_counts()
```

```python
police_incidents_pdf['Date Occurred'].value_counts().sort_index().plot(figsize=(20,10));
```
